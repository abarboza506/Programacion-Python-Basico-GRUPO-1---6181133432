#### **Paso 2: Recorrer los Datos de los Pacientes**

#### **Paso 3: Verificar si la Enfermedad ya Está en la Lista**

1. **Objetivo:** Verificar si la enfermedad ya fue registrada.
    
2. **Instrucción:**
    
    - Usa un `if` para verificar si la `enfermedad` ya está en `lista_enfermedades`.
    - Si **no** está en `lista_enfermedades`, agrega la enfermedad a `lista_enfermedades` y añade `1` a `lista_cantidades` para contarla por primera vez.
    - Si **sí** está en `lista_enfermedades`, encuentra la posición (`indice`) de la enfermedad en `lista_enfermedades` y aumenta en `1` el valor en `lista_cantidades` en la misma posición.

````python
# Pista: Usa `in` para verificar si la enfermedad ya está en lista_enfermedades.
# Pista: Si la enfermedad no está, agrégala a lista_enfermedades y empieza el conteo en lista_cantidades.
# Pista: Si la enfermedad ya está, usa `index` para encontrar su posición en lista_enfermedades y suma 1 en lista_cantidades en esa posición.

#enfermedad
#lista_enfermedades
#lista_cantidades

lista_enfermedades = []
lista_cantidades = []

for i in enfermedad:
    if i in lista_enfermedades:

        indice = lista_enfermedades.index(i)
        lista_cantidades[indice] += 1

    else:
        lista_enfermedades.append(i)
        lista_cantidades.append(1)

print(lista_enfermedades)
print(lista_cantidades)
````

#### **Paso 4: Imprimir el Reporte Final**

1. **Objetivo:** Mostrar el número de pacientes con cada enfermedad.
    
2. **Instrucción:**
    
    - Usa un bucle `for` para recorrer la lista `lista_enfermedades`.
    - En cada iteración, imprime la enfermedad y el número de pacientes (que está en `lista_cantidades` en la misma posición).

````python
#print("Reporte de enfermedades tratadas:")

#Titulos
print(f"{'Enfermedad':<20} {' '} {'Cantidad de Pacientes'}")
print("-" * 45)

# Imprimir tabla vertical
for enfermedad, cantidad in zip(lista_enfermedades, lista_cantidades):
    print(f"{enfermedad:<20} {'|'} {'         '} {cantidad}")
    print("-" * 45)

# Sumar los elementos de lista_cantidades
total = sum(lista_cantidades)

# Imprimir la suma total al final de la tabla
print(f"{'Total':<20} {'|'} {'         '} {total}")
print("-" * 45)
````

````python
# Pedimos las identificaciones de los dos pacientes
id_paciente1 = input("Ingrese la identificación del primer paciente: ")
id_paciente2 = input("Ingrese la identificación del segundo paciente: ")

# Función para buscar al paciente por identificación
def buscar_paciente(id_paciente):
    for paciente in datos_pacientes:
        if paciente[0] == id_paciente:
            return paciente
    return None


# Buscar los registros de los pacientes
paciente1 = buscar_paciente(id_paciente1)
paciente2 = buscar_paciente(id_paciente2)

try:

    #Impresión del reporte

    print(f"{'Paciente':<13} {'|'} {'Enfermedad':<13} {'|'} {'Medicamento':<13}")
    print("-" * 45)
    print(f"{paciente1[0]:<13} {'|'} {paciente1[4]:<13} {'|'} {paciente1[5]:<13}")
    print("-" * 45)
    print(f"{paciente2[0]:<13} {'|'} {paciente2[4]:<13} {'|'} {paciente2[5]:<13}")
    print("-" * 45)


    # Verificar si ambos pacientes existen en los datos
    print('\nResumen Pacientes:')
    print("-" * 18)

    #Comparación
    if paciente1 and paciente2:
        
        # Comparación enfermedad
        if paciente1[4] == paciente2[4]:
            print("\n* Ambos pacientes tienen la misma enfermedad.")
        else:
            print("\n* Los pacientes tienen enfermedades diferentes.")
        
        # Comparación medicamentos
        if paciente1[5] == paciente2[5]:
            print("\n* Ambos pacientes están tomando el mismo medicamento.")
        else:
            print("\n* Los pacientes están tomando medicamentos diferentes.")
    else:
        print("\nUno o ambos pacientes no se encontraron en la base de datos.")

except ValueError as ve:
    print(f"\nError: {ve}")

except Exception as e:
    print(f"\nOcurrió un error inesperado: {e}")

#"""
#Prueba con 702370234 - 215487963
#Salida esperada
#Enfermedades en común entre 702370234 y 215487963: ['migraña']
#Medicamentos en común entre 702370234 y 215487963: ['acetaminofen']
#"""
````
